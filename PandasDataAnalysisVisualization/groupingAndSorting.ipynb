```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.loc[
    reviews.groupby(['country', 'province'])['points'].idxmax()
]
```

```python
# First, find the maximum points per group
max_points = reviews.groupby(['country', 'province'])['points'].transform('max')
max_points
```

```python
# Then filter only rows where points == max_points
top_reviews = reviews[reviews['points'] == max_points]
top_reviews
```
