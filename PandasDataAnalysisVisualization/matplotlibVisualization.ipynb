```python
# Matplotlib Adding Grid Lines

# Add Grid Lines to a Plot
# With Pyplot, you can use the grid() function to add grid lines to the plot.


import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([10, 20, 30, 40])

# First plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot in position 1
plt.plot(x, y1)
plt.title("Plot 1")

# Second plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot in position 2
plt.plot(x, y2)
plt.title("Plot 2")

# Second plot
plt.subplot(2, 2, 3)  # 1 row, 2 columns, plot in position 2
plt.plot(x, y2)
plt.title("Plot 3")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)  # Row 1, Column 1
plt.subplot(2, 2, 2)  # Row 1, Column 2
plt.subplot(2, 2, 3)  # Row 2, Column 1
plt.subplot(2, 2, 4)  # Row 2, Column 2
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x, y2)
plt.title("Plot 3")

# this will prevent the overlapping 
plt.tight_layout() 

plt.show()
```

```python
# Colors
# You can set the color of each wedge with the colors parameter.

# The colors parameter, if specified, must be an array with one value for each wedge:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
# Legend
# To add a list of explanation for each wedge, use the legend() function:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
