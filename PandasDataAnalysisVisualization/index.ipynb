```python
# when you use methods like .isnull(), .fillna(), .dropna(), or even .set_index(), the changes do not automatically modify the original DataFrame unless you:



df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [80, 90, 70]
})

# Add 10 bonus points to each score
df['score'] = df['score'].apply(lambda x: x + 10)
df
```

```python
# When used on a DataFrame
# You can apply a function row-wise or column-wise by using the axis argument:

# axis=0: Apply to each column

# axis=1: Apply to each row

# Combine 'name' and 'score' into a single string per row
df.apply(lambda row: f"{row['name']} scored {row['score']}", axis=1)
```

```python
df
```

```python
df['summary'] = df.apply(lambda row: f"{row['name']} scored {row['score']}", axis=1)
df
```

```python
# Convert scores to grade letters
def grade(x):
    if x >= 90:
        return 'A'
    elif x >= 80:
        return 'B'
    elif x >= 70:
        return 'C'
    else:
        return 'D'

df['grade'] = df['score'].apply(grade)
df
```

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```
